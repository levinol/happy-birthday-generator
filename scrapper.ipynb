```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd 
```

```python
def fast_scrap(url, pagenumb = False, stih_tag = False):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    onePageTexts = []
    for elem in soup.find_all(class_='sfst'):
        if stih_tag:
            msg = elem.contents
            for idx, item in enumerate(msg):
                if str(item) == '<br/>':
                    msg[idx] = '\n'
            try:
                msg = ' '.join(msg)
                onePageTexts.append(msg.replace(u'\xa0', u' ').encode("utf-8"))
            except TypeError:
                print(url, msg)
        else:
            msg = elem.getText()
            onePageTexts.append(msg.replace(u'\xa0', u' ').encode("utf-8"))
    if pagenumb == True:
        for elem in soup.find_all(class_='allp'):
            maybe_number = elem.find('b', class_="curpg").getText()
            if maybe_number.isnumeric():
                break
        try:
            page_count = int(maybe_number)//len(onePageTexts)
            return np.array(onePageTexts), page_count + 1 if int(maybe_number)%len(onePageTexts) != 0 else page_count
        except ValueError:
            print("Не найдено число страниц")
            return np.array(onePageTexts) 
    else:
        return np.array(onePageTexts)
```

```python
Proza_texts, pgnmb = fast_scrap('http://pozdravok.ru/pozdravleniya/den-rozhdeniya/proza.htm', pagenumb = True)
for id in range(2, pgnmb+1):
    cur_url = f'http://pozdravok.ru/pozdravleniya/den-rozhdeniya/proza-{id}.htm'
    Proza_texts = np.append(Proza_texts, fast_scrap(cur_url))
```

```python
Stih_texts, pgnmb = fast_scrap('http://pozdravok.ru/pozdravleniya/den-rozhdeniya/stih.htm', pagenumb = True, stih_tag = True)
for id in range(2, pgnmb+1):
    cur_url = f'http://pozdravok.ru/pozdravleniya/den-rozhdeniya/stih-{id}.htm'
    Stih_texts = np.append(Stih_texts, fast_scrap(cur_url, stih_tag = True))
```

```python
pd.DataFrame(Proza_texts).to_csv("proza.csv")
pd.DataFrame(Stih_texts).to_csv("stih.csv")
```

```python
df_proza = pd.read_csv('proza.csv', names = ['N', 'text'], header = 0)  
df_stih =pd.read_csv('stih.csv', names = ['N', 'text'], header = 0)  
```

```python
import io
import ast
with io.open("proza.txt", 'w', encoding="utf-8") as f:
    for index, row in df_proza.iterrows():
        f.write('<s>Тип: Проза\nТекст: ' + ast.literal_eval(row.text).decode("utf-8") + '</s>\n')
with io.open("stih.txt", 'w', encoding="utf-8") as f:
    for index, row in df_stih.iterrows():
        f.write('<s>Тип: Стих\nТекст: ' + ast.literal_eval(row.text.replace('\n','\g')).decode("utf-8") + '</s>\n')
```

```python
from sklearn.model_selection import train_test_split

Proza_train, Proza_test = train_test_split(df_proza, test_size = 0.2, random_state=124)
Stih_train, Stih_test = train_test_split(df_stih, test_size = 0.2, random_state=124)
```

```python
with io.open("train.txt", 'w', encoding="utf-8") as f:
    for index, row in Proza_train.iterrows():
        f.write('<s>Тип: Проза\nТекст: ' + ast.literal_eval(row.text).decode("utf-8") + '</s>\n')
    for index, row in Stih_train.iterrows():
        f.write('<s>Тип: Стих\nТекст: ' + ast.literal_eval(row.text).decode("utf-8").replace('\n','\g') + '</s>\n') 
        
with io.open("valid.txt", 'w', encoding="utf-8") as f:
    for index, row in Proza_test.iterrows():
        f.write('<s>Тип: Проза\nТекст: ' + ast.literal_eval(row.text).decode("utf-8") + '</s>\n')
    for index, row in Stih_test.iterrows():
        f.write('<s>Тип: Стих\nТекст: ' + ast.literal_eval(row.text).decode("utf-8").replace('\n','\g') + '</s>\n')
```
